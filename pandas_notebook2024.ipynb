# pandas 7)
## using merge to merge two data frames

```python
df6 = pd.DataFrame({
    'Customer_ID': [1, 2, 3, 4, 5],
    'Name': ['John', 'Anna', 'Mike', 'Sara', 'Tom'],
    'Country': ['USA', 'Canada', 'USA', 'UK', 'Australia'],
    'Age': [28, 34, 29, 22, 41]
})
df6
```

```python
df7 =pd.DataFrame({
    'Customer_ID': [1,5,8,9,6],
    'Product': ['Laptop', 'Tablet', 'Phone', 'Monitor', 'Keyboard'],
    'Quantity': [1, 3, 2, 1, 5],
    'Price_per_unit': [1000, 300, 500, 200, 50]
})
df7
```

```python
#uisng merge function to merge the data -- merge function commonly working like joins in sql(inner,outer,left,right)
#here used inner join

df_new=pd.merge(df6,df7,on='Customer_ID',how='inner',indicator=True)
df_new
```

```python
#outer

df_new=pd.merge(df6,df7,on='Customer_ID',how='outer',indicator=True)
df_new
```

```python
#left
df_new=pd.merge(df6,df7,on='Customer_ID',how='left',indicator=True)
df_new
```

```python
#right

df_new=pd.merge(df6,df7,on='Customer_ID',how='right',indicator=True)
df_new
```

# Pandas 7) pivot tables
## pivot and pivot_tabel 

```python
import pandas as pd
from datetime import datetime, timedelta

# Generate sample date range
dates = [datetime(2024, 1, 1) + timedelta(days=i) for i in range(10)]

# Sample data
data = {
    'Date': dates,
    'Product': ['Laptop', 'Tablet', 'Phone', 'Monitor', 'Keyboard', 'Laptop', 'Mouse', 'Tablet', 'Phone', 'Monitor'],
    'Quantity_Sold': [5, 3, 8, 2, 10, 7, 12, 5, 6, 4],
    'Revenue': [5000, 900, 4000, 400, 500, 3500, 120, 1500, 3000, 800]
}

df = pd.DataFrame(data)
df
```

```python
#here i used pivot table

new=df.pivot(index='Date',columns='Product',values='Revenue')
new
```

```python
# pivot table is used to summaize and aggregate data inside dataframe

#using pivot tabel

new1=df.pivot_table(index='Date',columns='Product',aggfunc='sum')  #using the aggfunc method
new1
```

```python
#using grouper function 


new1=df.pivot_table(index=pd.Grouper(freq='N',key='Date'),columns='Product')  
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Generate sample date range
dates = [datetime(2024, 1, 1) + timedelta(days=i) for i in range(10)]

# Sample data
data = {
    'Date': dates,
    'Product': ['Laptop', 'Tablet', 'Phone', 'Monitor', 'Keyboard', 'Laptop', 'Mouse', 'Tablet', 'Phone', 'Monitor'],
    'Quantity_Sold': [5, 3, 8, 2, 10, 7, 12, 5, 6, 4],
    'Revenue': [5000, 900, 4000, 400, 500, 3500, 120, 1500, 3000, 800]
}

df = pd.DataFrame(data)
df
```

```python
type(df.Date[0])
```

```python
df.resample('Y').mean()

df['date'] = pd.to_datetime(df['date'])  # Convert 'date' column to datetime
```
